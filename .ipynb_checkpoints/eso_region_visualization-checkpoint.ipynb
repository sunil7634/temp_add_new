```python
# Magic Functions Go Here
%matplotlib inline
# Imports
import numpy as np
import astropy.io.fits as fits
import matplotlib.pylab as plt
from astropy.wcs import WCS
from astropy.visualization import (LogStretch, ManualInterval,
                                   ImageNormalize)
```

```python
data,header = fits.getdata('F475W_drc_chip_cleaned.fits',header=True)
wcs = WCS(header)
```

```python
norm = ImageNormalize(data, interval=ManualInterval(vmin=-0.0052, vmax=0.22), #zscale as in ds9
                      stretch=LogStretch(1))
```

## regions 

```python
import matplotlib as mpl
# MPL Setup
GOLD = (1 + np.sqrt(5))/2
mpl.rc('font', **{'size': 20})
mpl.rc('figure', **{'figsize': (20, 20/GOLD)})
```

```python
import pyregion

region_name='/Users/sunil/Documents/TestESO_137_002/ESO-137-002/09_color_mag_regions/test.reg'
r = pyregion.open(region_name)
```

```python
# Create the Figure
fig = plt.figure(figsize=(10,10))
ax = plt.subplot(projection=wcs)  # Assign the WCS projection
for t in ax.coords:
    t.display_minor_ticks(True)
    t.set_minor_frequency(10)

ax.tick_params(which='major', length=7,direction='in')
im=plt.imshow(data, norm=norm, cmap='gray_r', origin="lower")
plt.xlabel('Right Ascension (J2000)')
plt.ylabel('Declination (J2000)')
ax.coords[1].set_ticklabel(rotation=45)

patch_list, artist_list = r.get_mpl_patches_texts()
for p in patch_list:
    ax.add_patch(p)
for t in artist_list:
    ax.add_artist(t)
    
plt.savefig('eso_region.eps')    
plt.show()
```
