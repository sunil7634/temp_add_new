### Functions

```python
R=[1,2,3,4,5,6,7,8]
```

```python
def circum(r):
    return 2*np.pi*r
def area(r):
    return np.pi * r**2
```

```python
circum(3)
```

```python
for i in range(5):
    print(circum(i))
```

```python
for item in R:
    print(area(item))
```
